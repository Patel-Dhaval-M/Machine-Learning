```python
def simple_linear_regression(input_feature, output):
    meanOfXMuliplyY = (input_feature * output).mean()
    meanOfX = input_feature.mean()
    meanOfY = output.mean()
    meanOfXSquare = (input_feature ** 2).mean()
    numerator = meanOfXMuliplyY - (meanOfX*meanOfY)
    denominator = meanOfXSquare - (meanOfX*meanOfX)
    slope = numerator/float(denominator)
    intercept = meanOfY - (slope*meanOfX)
    return(intercept, slope)
```

```python
intercept_model, slope_model = simple_linear_regression(house_train['sqft_living'],house_train['price'])
```

```python
def get_regression_predictions(input_feature, intercept, slope):
    predicted_output = intercept + (slope * input_feature)
    return(predicted_output)
```

```python
house_price = get_regression_predictions(2650,intercept_model,slope_model)
```

```python
print house_price
```

```python
def get_residual_sum_of_squares(input_feature, output, intercept,slope):
    RSS = ((output - (intercept + (slope * input_feature))) ** 2).sum()
    return(RSS)
```

```python
RSS_train = get_residual_sum_of_squares(house_train['sqft_living'], house_train['price'],intercept_model, slope_model)
```

```python
RSS_test = get_residual_sum_of_squares(hous_test['sqft_living'], hous_test['price'], intercept_model, slope_model)
```

```python
print RSS_test
print RSS_train
```

```python
def inverse_regression_predictions(output, intercept, slope):
    estimated_input = (output - intercept)/float(slope)
    return(estimated_input)
```

```python
sqft_living_input = inverse_regression_predictions(800000, intercept_model, slope_model)
```
